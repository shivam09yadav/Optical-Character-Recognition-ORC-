# **Importing the libraries:**

```python
import keras_ocr
import matplotlib.pyplot as plt
```

# **Download Weights**

Calling the pipeline function from the keras_ocr.pipeline class and saving into a variable called as pipeline.
In this code keras_ocr will automatically download pre trained weights for the detector and recognizer and save it into the variable pipeline.

```python
pipeline = keras_ocr.pipeline.Pipeline()
```

# **Get the Images that we want to extract characters:**

```python
images = [
    keras_ocr.tools.read(img) for img in ['image1.jpg',
                                          'image3.jpg'
    ]
]
```

Validating the Images

```python
len(images)
```

Loading the Images

## **Get the predictions:**

1.Calling the recognize command and passing the two images.
               2.Saving it in the variable called as prediction_groups

Now each list of predictions in the prediction_groups will essentially have a word,box tuple.
so word would mean that what is the text that is recognized and box will contain the bounding box of where that word has been recognized.
This would be saved in the variable called prediction_groups.

```python
prediction_groups = pipeline.recognize(images)
```

Making a prediction and also annotating the orignal images with the text along with where the text have been discovered.

```python
fig, axs = plt.subplots(nrows=len(images), figsize=(10, 20))
for ax, image, predictions in zip(axs, images, prediction_groups):
    keras_ocr.tools.drawAnnotations(image=image, 
                                    predictions=predictions, 
                                    ax=ax)
```
